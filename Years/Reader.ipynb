```python
import pandas as pd
```

```python
df = pd.read_csv("years")
df.head()

```

```python
movie = list(df.movie)
year = list(df.year)

movieYearDictionary = {}
for x,y in zip(movie, year):
    movieYearDictionary[x] = y
movieYearDictionary
```

```python
df2 = pd.read_csv("data")
df2.head()
```

```python
movieNames = list(df2.movieName)
movieYears = []
for myMovie in movieNames:
    try:
        year =  movieYearDictionary[myMovie] 
        movieYears.append(year)
    except:
        print("Not found ", myMovie)
        movieYears.append(0)
        
        
```

```python
len(movieYears), len(movieNames)
```

```python
df3 = df2.copy()
df3["years"] = movieYears
```

```python
df3
```

```python
df3.to_csv("moviesWithYears.csv", header = True, index = False, columns = df3.columns)
```
